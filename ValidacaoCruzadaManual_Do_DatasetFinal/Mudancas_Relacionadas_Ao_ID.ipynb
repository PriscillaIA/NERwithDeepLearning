O objetivo deste jupyter notebook é criar uma nova coluna no dataset_rotulado, chamada Documentos, a qual contém referencia a qual documentos (páginas) extraída do sites de venda de bebida cada token pertence
Também tem como objetivo remover os id na coluna Palavras, para que os ids não sejam visto como sentenças.

```python
import pandas as pd
```

```python
## ler o Dataset
dataset = pd.read_csv("./Dataset_Padrao_IOB_Versao4.csv", usecols=['Palavras', 'Rotulo', 'Sentenca','Inicio','Fim'], encoding='utf-8-sig')
```

```python
## Exibi Dataset
display(dataset)
```

```python
## Cria uma nova coluna chamada 'id'
dataset['Documento'] = None
```

```python
## Exibi Dataset
display(dataset)
```

```python
## Pega os index's onde na coluna 'Palavras' o texto é 'id', salva em uma lista
lista_index_ids = list(dataset.loc[dataset['Palavras'] == 'id'].index)
## Adiciona o tamanho do dataset ao final da lista com index's (para q a ideia usada no for abaixo funcione)
lista_index_ids.append(len(dataset))
```

```python
## Realiza um laço para adicionar na coluna 'id' criada o id correspondente, adicionando via intervalo entre cada 'id'. 
for cont in range(len(lista_index_ids)-1):
    dataset[lista_index_ids[cont]:lista_index_ids[cont+1]]['Documento'] = str(cont+1)
```

```python
## Exibi Dataset
display(dataset)
```

```python
## Remove o tamanho do dataset ao final da lista com index's
lista_index_ids.remove(len(dataset))
```

```python
## Criando a lista quq armazena o index dos valores a serem removidos
lista_remove = []
for index in lista_index_ids:
    lista_remove.append(index)
    lista_remove.append(index+1)
    lista_remove.append(index+2)
```

```python
## Retira do dataset os valores indesejados
dataset = dataset.drop(axis=0, index=lista_remove)
```

```python
## Resentado os valores da colua index, após os ids terem sido removidos
dataset = dataset.reset_index()
dataset = dataset.drop(columns='index')
```

```python
## Atualiza os números do campo Sentença, pois após a remoção dos ids os números não ficaram ordenados.
dataset.loc[:, "Sentenca"] = dataset["Sentenca"].apply(lambda x: x - 1)
```
